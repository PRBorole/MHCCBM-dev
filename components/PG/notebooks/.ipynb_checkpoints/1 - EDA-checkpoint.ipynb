```python
# Read and create data
PG_df = pd.read_csv('./../../../data/PG/Data_S5.csv')
PG_df = PG_df.drop_duplicates('peptide')[['n_flank','c_flank','peptide','hit']].reset_index(drop=True)
PG_df['1flank'] = [n[-1]+p+c[0] for n,p,c in zip(PG_df['n_flank'], PG_df['peptide'], PG_df['c_flank'])]
PG_df['2flank'] = [n[-2:]+p+c[0:2] for n,p,c in zip(PG_df['n_flank'], PG_df['peptide'], PG_df['c_flank'])]
PG_df['3flank'] = [n[-3:]+p+c[0:3] for n,p,c in zip(PG_df['n_flank'], PG_df['peptide'], PG_df['c_flank'])]
PG_df['4flank'] = [n[-4:]+p+c[0:4] for n,p,c in zip(PG_df['n_flank'], PG_df['peptide'], PG_df['c_flank'])]
PG_df['5flank'] = [n[-5:]+p+c[0:5] for n,p,c in zip(PG_df['n_flank'], PG_df['peptide'], PG_df['c_flank'])]
# PG_df.to_csv('./../../../data/PG/PG.csv')
PG_df
```
