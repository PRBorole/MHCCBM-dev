```python
# read training data
PS_benchmark_df = pd.read_csv('./results/PS_benchmark.csv', index_col=0)

# TD proba
min_max_scaler = preprocessing.MinMaxScaler()
PS_benchmark_df['TD_proba'] = min_max_scaler.fit_transform(PS_benchmark_df[['MFI_ratio']]).reshape(-1)
PS_benchmark_df
```

```python
# keep only relevant
PS_benchmark_df = PS_benchmark_df[['peptide',
                                 'allele',
                                 'PG_proba',
                                 'TAP_proba',
                                 'BA_proba',
                                 'TD_proba',
                                 'PG_y',
                                 'TAP_y',
                                 'BA_y',
                                 'TD_y',
                                 'PS_proba',
                                 'PS_y',
                                 'hit']]
```

```python
# Create X and y
X_y_df = PS_benchmark_df[['peptide', 
                           'allele', 
                           'PG_proba',
                            'TAP_proba',
                            'BA_proba',
                            'TD_proba',
                            'PS_proba',
                            'PS_y',
                            'hit']].dropna()

X = X_y_df[['PG_proba',
            'TAP_proba',
            'BA_proba',
            'TD_proba']].to_numpy()

y = X_y_df['hit'].to_numpy()
```

```python
# load model
with open('./models/PS_SOTA_clf_proba.pkl', 'rb') as f:
    clf = pickle.load(f)
f.close()

clf
```

```python
# Make predictions
y_pred = clf.predict(X)
y_proba_pred = clf.predict_proba(X)

X_y_df['MHCCBM_proba'] = y_proba_pred[:,1]
X_y_df['MHCCBM_y'] = y_pred

AUROC = roc_auc_score(y, y_proba_pred[:,1])
print("AUROC: ", AUROC)

# calculate precision-recall curve
precision, recall, thresholds = precision_recall_curve(y, y_proba_pred[:,1])
AUPRC = auc(recall, precision)
print("AUPRC: ", AUPRC)

# calculate f1 score
f1 = f1_score(y, y_pred)
print("f1: ", f1)

# calculate accuracy score
accuracy = accuracy_score(y, y_pred)
print("accuracy: ", accuracy)

# calculate precision score
precision = precision_score(y, y_pred)
print("precision: ", precision)

# calculate recall score
recall = recall_score(y, y_pred)
print("recall: ", recall)


```

## Calculate Allele-wise


```python
allele_ls = X_y_df['allele'].unique()
result_df = {'allele':[],
             'MHCCBM_AUROC':[],
            'MHCCBM_AUPRC':[],
            'MHCCBM_F1':[],
            'MHCCBM_Accuracy':[],
            'MHCCBM_Precision':[],
             'MHCCBM_Recall':[],
            'MHCflurry_AUROC':[],
            'MHCflurry_AUPRC':[],
            'MHCflurry_F1':[],
            'MHCflurry_Accuracy':[],
            'MHCflurry_Precision':[],
             'MHCflurry_Recall':[]}


for allele in allele_ls:
    allele_df = X_y_df[X_y_df['allele']==allele]
    result_df['allele'] = result_df['allele'] + [allele]
    
    
    ######### MHCCBM
    y = allele_df['hit']
    y_pred = allele_df['MHCCBM_y']
    y_proba_pred = allele_df['MHCCBM_proba']
    
    # Calculate AUROC
    AUROC = roc_auc_score(y, y_proba_pred)
    result_df['MHCCBM_AUROC'] = result_df['MHCCBM_AUROC'] + [AUROC]

    # calculate precision-recall curve
    precision, recall, thresholds = precision_recall_curve(y, y_proba_pred)
    AUPRC = auc(recall, precision)
    result_df['MHCCBM_AUPRC'] = result_df['MHCCBM_AUPRC'] + [AUPRC]

    # calculate f1 score
    f1 = f1_score(y, y_pred)
    result_df['MHCCBM_F1'] = result_df['MHCCBM_F1'] + [f1]

    # calculate accuracy score
    accuracy = accuracy_score(y, y_pred)
    result_df['MHCCBM_Accuracy'] = result_df['MHCCBM_Accuracy'] + [accuracy]

    # calculate precision score
    precision = precision_score(y, y_pred)
    result_df['MHCCBM_Precision'] = result_df['MHCCBM_Precision'] + [precision]

    # calculate recall score
    recall = recall_score(y, y_pred)
    result_df['MHCCBM_Recall'] = result_df['MHCCBM_Recall'] + [recall]
    
    ######### MHCflurry
    y = allele_df['hit']
    y_pred = allele_df['PS_y']
    y_proba_pred = allele_df['PS_proba']
    
    # Calculate AUROC
    AUROC = roc_auc_score(y, y_proba_pred)
    result_df['MHCflurry_AUROC'] = result_df['MHCflurry_AUROC'] + [AUROC]

    # calculate precision-recall curve
    precision, recall, thresholds = precision_recall_curve(y, y_proba_pred)
    AUPRC = auc(recall, precision)
    result_df['MHCflurry_AUPRC'] = result_df['MHCflurry_AUPRC'] + [AUPRC]

    # calculate f1 score
    f1 = f1_score(y, y_pred)
    result_df['MHCflurry_F1'] = result_df['MHCflurry_F1'] + [f1]

    # calculate accuracy score
    accuracy = accuracy_score(y, y_pred)
    result_df['MHCflurry_Accuracy'] = result_df['MHCflurry_Accuracy'] + [accuracy]

    # calculate precision score
    precision = precision_score(y, y_pred)
    result_df['MHCflurry_Precision'] = result_df['MHCflurry_Precision'] + [precision]

    # calculate recall score
    recall = recall_score(y, y_pred)
    result_df['MHCflurry_Recall'] = result_df['MHCflurry_Recall'] + [recall]

    
result_df = pd.DataFrame(result_df)
result_df['diff_AUROC'] = result_df['MHCCBM_AUROC'] - result_df['MHCflurry_AUROC']
result_df['diff_AUPRC'] = result_df['MHCCBM_AUPRC'] - result_df['MHCflurry_AUPRC']
result_df['diff_F1'] = result_df['MHCCBM_F1'] - result_df['MHCflurry_F1']
result_df['diff_Precision'] = result_df['MHCCBM_Precision'] - result_df['MHCflurry_Precision']
result_df['diff_Recall'] = result_df['MHCCBM_Recall'] - result_df['MHCflurry_Recall']
result_df
```

```python
fontsize = 14
fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(5,5))

plot_df = result_df[['MHCflurry_Precision', 'MHCCBM_Precision','allele']].melt(['allele'])
# plot_df = result_df[['MHCflurry_AUPRC', 'MHCCBM_AUPRC','allele']].melt(['allele'])


b = sns.lineplot(data=plot_df, x='variable', y='value',
                 ax=axs,
                 legend=False)

# b.set_xticklabels(b.get_xticklabels(), fontsize=fontsize-5, rotation=90)
# b.set_yticklabels(b.get_yticklabels(), fontsize=fontsize-5)
# b.set_ylabel('')
b.set_xlabel('')
sns.despine(top=True, right=True)
```
